## Scikit-learn fit

```python
from sklearn.cluster import KMeans

KMeans_clst = KMeans(n_clusters=3)
KMeans_clst.fit(X)
```

```python
centroids = KMeans_clst.cluster_centers_
y = KMeans_clst.labels_
```

```python
centroids, y
```

```python
from matplotlib.colors import ListedColormap

fig, ax = plt.subplots(figsize=(6,6))
ax.scatter(X[:, 0], X[:, 1], s=20, c=y, cmap=ListedColormap(['r', 'g', 'b']))
ax.scatter(centroids[:, 0], centroids[:, 1], s=50, c='k', marker='P')
ax.set_title('K-means Data Set 2')
ax.set_xlabel('X1')
ax.set_ylabel('X2')
None
```

# DBSCAN
Tackle the weaknesses of K-means algorithms.

```python
from sklearn.cluster import DBSCAN

DBSCAN_clst = DBSCAN()
DBSCAN_clst.fit(X)
```

```python
y2 = DBSCAN_clst.labels_
y2
```

```python
fig, ax = plt.subplots(figsize=(6,6))
ax.scatter(X[:, 0], X[:, 1], s=20, c=y2, cmap=ListedColormap(['k', 'r', 'g', 'b']))
ax.set_title('DBSCAN Data Set 2')
ax.set_xlabel('X1')
ax.set_ylabel('X2')
None
```

```python
from sklearn.datasets import make_blobs

X, y = make_blobs(n_samples=1000, random_state=150)
transformation = [[0.60834549, -0.63667341], [-0.40887718, 0.85253229]]
X_aniso = np.dot(X, transformation)
y_pred_1 = KMeans(n_clusters=3).fit_predict(X_aniso)
y_pred_2 = DBSCAN(eps=0.2).fit_predict(X_aniso)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(14,6))
ax[0].scatter(X_aniso[:, 0], X_aniso[:, 1], s=20, c=y_pred_1)
ax[0].set_title('K-means Anisotropicly')
ax[0].set_xlabel('X1')
ax[0].set_ylabel('X2')
ax[1].scatter(X_aniso[y_pred_2!=-1, 0], X_aniso[y_pred_2!=-1, 1], s=20, c=y_pred_2[y_pred_2!=-1])
ax[1].scatter(X_aniso[y_pred_2==-1, 0], X_aniso[y_pred_2==-1, 1], s=20, c='k')
ax[1].set_title('DBSCAN Anisotropicly')
ax[1].set_xlabel('X1')
ax[1].set_ylabel('X2')
None
```
